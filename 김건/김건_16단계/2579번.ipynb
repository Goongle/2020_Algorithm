```python
value = int(input())

ary = []
for i in range (0,value) :
    temp = int(input())
    ary.append(temp)
    
#         .
#       .
dp = []
if value == 1 :
    dp.append(ary[0]) # 0  
elif value == 2:
    dp.append(ary[0]) # 0  
    dp.append(ary[0] + ary[1]) # 1  
else:
    dp.append(ary[0]) # 0  
    dp.append(ary[0] + ary[1]) # 1  
    dp.append(max(ary[1]+ary[2],ary[2]+ary[0]))# 2  
    for i in range (3,value): 
        temp = max(dp[i-3] + ary[i-1] + ary[i],dp[i-2] + ary[i])
        dp.append(temp)
print(dp[value-1])
```
