## 거꾸로 생각하기 - Dynamic Programming


일반적으로 사람들은 이런식으로 구성이 된다고 생각을 한다. why? 2번 칸에 오르기 위해 그 이전 단계를 생각해야 하니까.
하지만 해당 문제는 이러한 접근 방식과 다르게 구성된다.

이러한 방식으로 내가 i 칸에 이미 올라왔다고 생각하고 뒤를 돌아보는 것. 즉, 풀어서 설명하면
Dp 라는 ary는 해당 칸에 올라오기 위한 Max 값을 저장한다.
이때, i 칸에 올라오는 루트는 바로 이전 칸(i-1)에서 올라오는 경우와 i-2 칸에서 올라오는 경우가 있다.
i-1 칸에서 올라오는 경우 i-2 칸에서 접근이 불가능하므로 i-3 칸에서 접근하였을 것이다.
i-2 칸에서 올라오는 경우 바로 접근이 가능하다.

따라서 아래와 같이 코드 구성을 하면 답을 구할 수 있다.

```python
value = int(input())

ary = []
for i in range (0,value) :
    temp = int(input())
    ary.append(temp)
    
# 한번에 두 계단 혹은 한 계단씩 오를 수 있다.
# 연속된 세 개의 계단을 밟는 것은 안된다.
dp = []
if value == 1 :
    dp.append(ary[0]) # 0 번째 칸
elif value == 2:
    dp.append(ary[0]) # 0 번째 칸
    dp.append(ary[0] + ary[1]) # 1 번째 칸
else:
    dp.append(ary[0]) # 0 번째 칸
    dp.append(ary[0] + ary[1]) # 1 번째 칸
    dp.append(max(ary[1]+ary[2],ary[2]+ary[0]))# 2 번째 칸
    for i in range (3,value): 
        temp = max(dp[i-3] + ary[i-1] + ary[i],dp[i-2] + ary[i])
        dp.append(temp)
print(dp[value-1])
```
