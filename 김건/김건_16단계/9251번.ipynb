## Dynamic Programming

두 문자열 간 최장거리문자열 구하기

문제는 즉 두 문자열을 구성할 수 있는 부분수열 중 가장 긴 것을 찾는 것이다. (이는 이전의 최장거리수열을 찾는 것이 아님!)

* 결국 답지를 보고 풀게 되었다.

이는 2차원 배열로 접근해야하는 문제이다.

즉, Ary[i][j] 를 볼 때, i는 S1의 단어 ex) ACKD 면 A C K D 
j는 S2의 단어 ex) CED 면 C E D 를 의미한다.

ex) ACKD 와 CED 가 있다고 가정하자
 A C K D  이런 식으로 배열이 구성될 것이다.
C
E
D

이렇게 될 경우 i에서 A와 j에서 C 가 들어가면 겹치는 것이 없으므로 0이 된다.
하지만 j 를 다 돌고 i 가 이동하여 C가 되었다고 가정하자. 그렇게 될 경우 i와 j의 C가 만나 1을 구성하게 된다. 이런식으로 반복하면 문제를 풀 수 있다. 


이처럼 2중 배열을 사용하여 최대 길이를 구하면 풀 수 있다. (사실 이 문제는 나도 잘 모르겠다 ㅠ )

```python
temp1 = input()
temp2 = input()
temp1_length = len(temp1)
temp2_length = len(temp2)
ary = []
for i in range (0,temp1_length +1 ) :
    temp = []
    for j in range (0,temp2_length + 1) :
        temp.append(0)
    ary.append(temp)
for i in range (1,temp1_length+1) :
    for j in range (1,temp2_length+1) :
        if temp1[i-1] == temp2[j-1] :
            ary[i][j] = ary[i-1][j-1] + 1
        else :
            temp = max(ary[i-1][j], ary[i][j-1])
            ary[i][j] = temp
print(ary[temp1_length][temp2_length])
```
