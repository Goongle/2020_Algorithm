## Dynamic Programming

DP를 이용하고 있으나 이중 배열을 사용하고 있어 쉽게 감이 잡히지 않았다. 
이 문제는 다음과 같다.

10 20 30 20 10 와 같이 숫자가 나열되어 있을 경우 가장 긴 거리의 배열이 어떻게 되는지 묻는 문제이다.

Brute Force를 통해 전부다 접근을 할 수도 있으나 Dp를 사용하면 더욱 간편하게 구성이 가능하다.



해당 그림을 참고해서 보면 ARP 배열은 원래의 숫자를 담은 것이고 DP 배열은 순차적으로 길이가 가장 긴 수열을 체크하기 위한 배열이다.

전체 로직을 보면 이중 포문을 돌고 있으며 핵심은 i인 나보다 작은 j들에 대해서 탐색하는 것이다.

30 20 10 20 30 이라는 숫자가 있다고 가정해보자. 그렇게 될 경우 처음엔 dp 를 모두 1로 초기화한다.
1  1  1  1  1 

이렇게 되고 순차적으로 비교한다. 

30부터 10까지 이보다 작은 수는 이전에 없으니 Pass

20의 관점에서 보면 20보다 작은 수인 10이 존재하므로 1 -> 2로 증가를 시켜줘야 한다. 이때 수식이
dp[i] = max(dp[i],dp[j] +1) 이렇게 되는데 이유는 바로
dp[j] + 1 즉 이전 수보다 큰 수의 배열이 존재할 경우 현재의 i를 껴서 수열의 길이를 완성하는 것이다.
하지만 이전의 수의 길이가 1로 동일하다면 연속된 배열이 아니므로 추가하지 않는 것이다.

이런식으로 구하면 최장 길이 수열을 구할 수 있다.

```python
value = int(input())
temp = input().split()
input_ary = []
for i in range (0,value) :
    num = int(temp[i])
    input_ary.append(num)
dp = [1] * value

for i in range (0,value) :
    for j in range (i,-1,-1) :
        if input_ary[i] > input_ary[j]:
            dp[i] = max(dp[i],dp[j]+1)
            
            
print(dp)
print(max(dp))
```
